## Análise dos dados

Os dados utilizados nesse projeto foram disponibilizados pela [LNB]('https://lnb.com.br/liga-ouro/'), e são referentes à temporada 2019 da Liga Ouro, divisão de acesso ao NBB.

Para facilitar nosso entendimento, vamos trabalhar cada categoria de dados que temos separadamente. Elas são:
* Estatísticas dos times da Liga Ouro 2019
* Estatísticas dos jogadores de todas as equipes
* Estatísticas dos jogos da equipe Cerrado Basquete na temporada 2019

```
# importando os pacotes necessários
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
```

```
# importando dados do time temporada 2019
df_ouro = pd.read_csv('https://raw.githubusercontent.com/rafaelnduarte/Cerrado_Basquete/master/data/Cerrado_stats.CSV', encoding='latin-1')

# importando dados de jogos / rodada 1 - turno 1 / 2019
df_t1 = pd.read_csv('https://raw.githubusercontent.com/rafaelnduarte/Cerrado_Basquete/master/data/cerrado_r1_t1.csv',encoding='latin-1')

# importando dados de jogos / rodada 1 - turno 2 / 2019
df_t2 = pd.read_csv('https://raw.githubusercontent.com/rafaelnduarte/Cerrado_Basquete/master/data/cerrado_r1_t2.csv', encoding='latin-1')

# imoprtando dados dos jogadores na temporada 2019
df_players = pd.read_csv('https://raw.githubusercontent.com/rafaelnduarte/Cerrado_Basquete/master/data/liga_ouro_players.csv', encoding='latin-1')

```

```
# creating the correlation matrix
df_corr = df_ouro.copy()
df_corr.drop('jo',axis=1, inplace=True)
df_corr.drop(['dd', 'td', 'id'], axis=1, inplace=True)
corr = df_corr.corr()

# plotting the heatmap
f, ax = plt.subplots(figsize=(12, 9))

sns.heatmap(corr, fmt='.2f', square=True, linecolor='white', vmax=1.0, annot=False);
plt.tight_layout()
```

```
# number of variables to be selected
k = 10

# finding the most correlated variables
cols = corr.nlargest(k, 'pts')['pts'].index
cm = np.corrcoef(df_corr[cols].values.T)

# plotting the heatmap
f, ax = plt.subplots(figsize=(12, 9))
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.tight_layout()
```

O senso comum do basquete nos diz que algumas estatísticas são mais importantes que outras.
É óbvio que a variável mais relacionada a vitória seria pontos. Entretanto, quais variáveis mais influenciam a variável média de pontos?

Quais seriam as primeiras que viriam a sua mente? Aproveitamento dos arremessos de quadra? 2 pontos, 3 pontos, assistências?

Esse gráfico conhecido como **Heatmap** (Mapa de calor, em tradução livre), nos mostra as 10 variáveis mais relacionadas à variável pontos.

>* ll_pct - Aproveitamento de lances livres
* rd - Rebotes defensivos
* ef - Eficiência
* rt - Rebotes totais
* br - Bolas recuperadas
* ro - Rebotes ofensivos
* llc - Lances livres convertidos
* as - Assistências
* 3p_pct - Aproveitamento de arremessos de 3 pontos


**Conclusões**

* O que chama mais atenção é a quantidade de variáveis defensivas estando altamente relacionadas com a média de pontos. Isso comprova a importância do trabalho defensivo para obter bons resultados ofensivos.

* Outra correlação interessante é em relação aos lances livres. Duas variáveis relacionadas a esse fundamento estão entre as mais importantes, sendo elas o aproveitamento de lances livres (ll_pct, que foi a mais relacionada) e a quantidade de lances livres convertidos (llc).

* A variável relacionada ao aproveitamento de arremessos de 3 pontos também é super importante, visto que é um aspecto marcante do basquete moderno, mostrando que essa tática também se faz presente no basquete nacional, a exemplo da NBA.

* A variável 'jo' não será tão relevante para nós no momento, pois ela se refere à quantidade de jogos em que o time participou. Ela poderá ser útil para outras análises.

**Conclusões acerca dos boxplots**

* A média de faltas cometidas é maior que a média de assistências por jogo.
* A média de erros totais é maior que a média de assistências por jogo.



**Comparações com o NBB**

Utilizando dados do site da LNB sobre o NBB, podemos também traçar alguns comparativos para as principais variáveis entre os times da Liga Ouro e NBB.

* O número de assistências por jogo parece surpreendentemente baixo na Liga Ouro. Para referência, a média de assistências por jogo do NBB é de 17.42, maior do que o valor máximo pra um time da Liga Ouro (15.00 / Campo Mourão).
* Quando comparados aos números do NBB, as porcentagens de aproveitamento de arremessos de 2 e 3 pontos são próximas.
* Os números de eficiência também são consideravelmente mais altos no NBB, tendo média de 90.62, contra 81.63 dos times da Liga Ouro.
* A média de rebotes totais na Liga Ouro (39.50) é marginalmente mais alta que a do NBB (37.40), que pode ser reflexo da maior eficiência dos times do NBB.
* Aproveitamento de Lances livres consideravelmente superior no NBB (73.11) em face da Liga Ouro (66.13).
* Média de bolas recuperadas marginalmente superior na Liga Ouro (7.05) em relação ao NBB (6.82).

#### Conclusões
A partir da análise deste dataset, conseguimos identificar quais as estatísticas que mais influenciam o total de pontos de um time, que leva à vitória das partidas.

Também identificamos as diferenças entre a performance dos times da Liga Ouro e do NBB, que é o objetivo dos times que disputam a Liga Ouro. Podemos destacar a grande diferença no número de assistências, aproveitamento de lances livres, e eficiência, bem como diferenças marginais em variáveis como rebotes totais e bolas recuperadas, que evidenciam o efeito da eficiência.
