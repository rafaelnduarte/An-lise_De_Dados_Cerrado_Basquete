```
# creating the correlation matrix
df_corr = df_ouro.copy()
df_corr.drop(['jo','dd', 'td', 'Id'], axis=1, inplace=True)
corr = df_corr.corr()

# plotting the heatmap
f, ax = plt.subplots(figsize=(12, 9))

sns.heatmap(corr, fmt='.2f', square=True, linecolor='white', vmax=1.0, annot=False);
plt.tight_layout()
```

Acima, podemos observar o nível de correlação entre as variáveis do dataset e a variável vitória.

Essa visualização conhecida como **Heatmap**, ou "Mapa de calor" em tradução livre faz um excelente trabalho em nos ajudar a visualizar essas correlações.

Entretanto, podemos melhorar consideravelmente essa visualização, focando apenas no mais importante.

```
# correlação Liga Ouro
k = 10

# encontrando as variáveis mais correlacionadas
cols = corr.nlargest(k, 'vit')['vit'].index
cm = np.corrcoef(df_corr[cols].values.T)

# plotando o heatmap
f, ax = plt.subplots(figsize=(12, 9))
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.tight_layout()
```

Agora podemos ver com mais clareza as variáveis que mais influenciam o resultado do jogo.

Algumas delas são "óbvias" e não são surpresa, como eficiência, arremessos tentados e convertidos de 2 pontos, rebotes defensivos e assistências.

Entretanto, nós estamos buscando detalhes que vão além do óbvio. Por isso, vamos avaliar mais de perto as outras variáveis, e entender como elas se relacionam com o resto das variáveis.

* Uma coisa que salta aos olhos é o fato da variável pontos não estar entre as mais relacionadas à vitória. Isso nos ajuda, pois seria óbvio. Entretanto, como pode ser que o aspecto que **define** o vencedor da partida não seja o mais correlacionado à vitória?
* A variável br_er nos mostra o índice entre bolas recuperadas e erros. Isto é, roubar mais bolas do que perder.
* A variável t_fc nos mostra a relação entre tocos e faltas. Conseguir uma boa média nesse aspecto é super importante, pois mais abaixo veremos que lances livres convertidos é uma das variáveis mais relacionadas à pontuação.
* A variável ia_pct é o índice de aproveitamento de assistências por erros. Mais uma vez, conseguimos confirmar que a variável mais importante realmente é eficiência. 

Agora que temos uma idéia dos aspectos que mais ajudam um time a vencer, vamos analisar que variáveis mais influenciam essas variáveis.

```
# analisando a variável eficiência
# number of variables to be selected
k = 11

# finding the most correlated variables
cols = corr.nlargest(k, 'ef')['ef'].index
cm = np.corrcoef(df_corr[cols].values.T)

# plotting the heatmap
f, ax = plt.subplots(figsize=(12, 9))
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.tight_layout()
```

Mais uma vez vemos uma boa representação de estatísticas defensivas. Vamos desconsiderar a variável vitória, pois já analisamos a relação dessa com a variável Eficiência.

É importante destacar:
* 5 das variáveis mais importantes para a eficiência são defensivas.
* Grande influência das estatísticas de rebotes (defensivos, ofensivos e totais)
* Bolas recuperadas em função de erros também tem grande papel em relação à eficiência.
* Previamente, tínhamos visto a importância da variável tocos por faltas no resultado do jogo. Aqui, além dela também ter papel importante, temos também a porcentagem de acerto dos lances livres, capitalizando nas faltas do adversário, e aproveitando "o arremesso mais fácil" do jogo.

Aqui, vamos destacar que **James Harden**, do Houston Rockets é atualmente o jogador que mais cava faltas e tenta lances livres, com aproveitamento na casa dos 80%. Isso também pode ser considerado efeito do trabalho de analytics feito pelo Houston Rockets, capitalizando nesse aspecto do jogo, utilizando bem as características de seus jogadores.

Agora, vamos analisar as variáveis que mais influenciam a variável pontos.



```
# analisando a variável pontos
# number of variables to be selected
k = 11

# finding the most correlated variables
cols = corr.nlargest(k, 'pts')['pts'].index
cm = np.corrcoef(df_corr[cols].values.T)

# plotting the heatmap
f, ax = plt.subplots(figsize=(12, 9))
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.tight_layout()
```

* ass - Assistências
* to - Tocos
