```
# creating the correlation matrix
df_corr = df_ouro.copy()
df_corr.drop(['jo','dd', 'td', 'Id'], axis=1, inplace=True)
corr = df_corr.corr()

# plotting the heatmap
f, ax = plt.subplots(figsize=(12, 9))

sns.heatmap(corr, fmt='.2f', square=True, linecolor='white', vmax=1.0, annot=False);
plt.tight_layout()
```

Acima, podemos observar o nível de correlação entre as variáveis do dataset e a variável vitória.

Essa visualização conhecida como **Heatmap**, ou "Mapa de calor" em tradução livre faz um excelente trabalho em nos ajudar a visualizar essas correlações.

Entretanto, podemos melhorar consideravelmente essa visualização, focando apenas no mais importante.

```
# correlação Liga Ouro
k = 10

# encontrando as variáveis mais correlacionadas
cols = corr.nlargest(k, 'vit')['vit'].index
cm = np.corrcoef(df_corr[cols].values.T)

# plotando o heatmap
f, ax = plt.subplots(figsize=(12, 9))
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.tight_layout()
```

Agora podemos ver com mais clareza as variáveis que mais influenciam o resultado do jogo.

Algumas delas são "óbvias" e não são surpresa, como eficiência, arremessos tentados e convertidos de 2 pontos, rebotes defensivos e assistências.

Entretanto, nós estamos buscando detalhes que vão além do óbvio. Por isso, vamos avaliar mais de perto as outras variáveis, e entender como elas se relacionam com o resto das variáveis.

* Uma coisa que salta aos olhos é o fato da variável pontos não estar entre as mais relacionadas à vitória. Isso nos ajuda, pois seria óbvio. Entretanto, como pode ser que o aspecto que **define** o vencedor da partida não seja o mais correlacionado à vitória?
* A variável br_er nos mostra o índice entre bolas recuperadas e erros. Isto é, roubar mais bolas do que perder.
* A variável t_fc nos mostra a relação entre tocos e faltas. Conseguir uma boa média nesse aspecto é super importante, pois mais abaixo veremos que lances livres convertidos é uma das variáveis mais relacionadas à pontuação.
* A variável ia_pct é o índice de aproveitamento de assistências por erros. Mais uma vez, conseguimos confirmar que a variável mais importante realmente é eficiência. 

Agora que temos uma idéia dos aspectos que mais ajudam um time a vencer, vamos analisar que variáveis mais influenciam essas variáveis.

```
# analisando a variável eficiência
# number of variables to be selected
k = 11
df_corr.drop(['vit'], axis=1, inplace=True)
corr = df_corr.corr()

# finding the most correlated variables
cols = corr.nlargest(k, 'ef')['ef'].index
cm = np.corrcoef(df_corr[cols].values.T)

# plotting the heatmap
f, ax = plt.subplots(figsize=(12, 9))
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.tight_layout()
```

Mais uma vez vemos uma boa representação de estatísticas defensivas. Vamos desconsiderar a variável vitória, pois já analisamos a relação dessa com a variável Eficiência.

É importante destacar:
* 5 das variáveis mais importantes para a eficiência são defensivas.
* Grande influência das estatísticas de rebotes (defensivos, ofensivos e totais)
* Bolas recuperadas em função de erros também tem grande papel em relação à eficiência.
* Previamente, tínhamos visto a importância da variável tocos por faltas no resultado do jogo. Aqui, além dela também ter papel importante, temos também a porcentagem de acerto dos lances livres, capitalizando nas faltas do adversário, e aproveitando "o arremesso mais fácil" do jogo.

Aqui, vamos destacar que **James Harden**, do Houston Rockets é atualmente o jogador que mais cava faltas e tenta lances livres, com aproveitamento na casa dos 80%. Isso também pode ser considerado efeito do trabalho de analytics feito pelo Houston Rockets, capitalizando nesse aspecto do jogo, utilizando bem as características de seus jogadores.

**Boxplot**

Boxplot, ou Diagrama de caixa, é uma maneira gráfica de representar a alteração dos dados de uma variável por meio de quartis. É especialmente útil para observar *Outliers*, que são unidades descrepantes em relação ao restante da distribuição.

Numa representação de Boxplot, como o da imagem abaixo, os *Outliers* são representados como pontos individuais além do ponto máximo, ou abaixo do ponto mínimo.

<center><img alt="Boxplots" width="30%" src="https://cdn.escolaedti.com.br/wp-content/uploads/2013/09/resumo-cinco-numeros-box-plot.jpg"></center>

No nosso caso, é importante observar esses dados e entender o que faz com que esses *outliers* estejam nessa posição.

Vamos à visualização:

```
# boxplots
f, ax = plt.subplots(figsize=(16, 8))
fig = sns.boxplot(data=df_corr)
fig.axis(ymin=-10, ymax=100);
plt.xticks(rotation=90);
```

## Advanced Stats

<center><img alt="Advanced Stats" width="30%" src="https://img.freepik.com/free-vector/basketball-online-isometric_1284-22086.jpg?size=626&ext=jpg"></center>

Nesta seção analisaremos o impacto dos **Advanced Stats** (Estatísticas Avançadas, em tradução livre) na perfomance dos times.


Desde a popularização de submétricas e a integração de análises avançadas em outros esportes, há formas mais eficientes de se medir a expectativa de sucesso de um time. Estas análises avançadas por jogo serão mostradas abaixo. Elas mostram diferenças consideráveis entre vitórias e derrotas ao longo da temporada. Para termos uma ideia melhor, e um modelo mais apurado, seria necessária a aquisição de mais dados, de temporadas passadas.

O cálculo dos Advanced Stats é feito da seguinte forma:

>* Possessions = 0.96 x (FGA + Turnovers + (0.475 x FTA) - Offensive Rebounds 
* Offensive Rating = 100 x (Score / Possessions)
* Defensive Rating = 100 x (Opponent's Score / Possessions) (Dados faltantes)
* Strength of Schedule = 100 x (Offensive Rating - Defensive Rating) (Dados Faltantes)
* Performance Impact Estimator (PIE) = Score + FGM + FTM - FGA - FTA + Defensive Rebounds + (0.5 * Offensive Rebounds) + Assists + Steals + (0.5 * Blocks) - PF - Turnovers (Score adaptado para média de pontos ('pts')
* Team Impact Estimator (TIE) = PIE / (PIE + Opponent's PIE) (Calculado time a time)
* Assist ratio = 100 * Assists / (FGA + (0.475 * FTA) + Assists + Turnovers))
* Turnover ratio = 100 * Turnovers / (FGA + (0.475 * FTA) + Assists + Turnovers)
* True Shooting % = 100 * Team Points / (2 * (FGA + (0.475 * FTA)))
* Effective FG% = (FGM + 0.5 * Threes Made) / FGA
* Free Throw rate = FTA / FGA
* Offensive Rebound % = Offensive Rebounds / (Offensive Rebounds + Opponent's Defensive Rebounds) (Calculado time a time)
* Defensive Rebound % = Defensive Rebounds / (Defensive Rebounds + Opponent's Offensive Rebounds) (Calculado time a time)
* Total Rebound % = (Defensive Rebounds + Offensive Rebounds) / (Defensive Rebounds + Offensive Rebounds + Opponent's Defensive Rebounds + Opponent's Offensive Rebounds) (Calculado time a time)

```
# criando Estatísticas Avançadas - Liga Ouro
df_ouro['fga'] = df_ouro.apply(lambda x: x['2pt'] + x['3pt'], axis=1)
df_ouro['fgm'] = df_ouro.apply(lambda x: x['2pc'] + x['3pc'], axis=1)
df_ouro['poss'] = df_ouro.apply(lambda x: 0.96 * (x['2pt'] + x['er'] + (0.45 * x['llt'] - x['ro'])), axis=1)
df_ouro['off_rtg'] = df_ouro.apply(lambda x: 100 * (x['pts'] / x['poss']), axis=1)
df_ouro['pie'] = df_ouro.apply(lambda x: x['pts'] + x['fgm'] + x['llc'] - x['fga'] - x['llt'] + x['rt'] + (0.5 * x['ro']) + x['as'] + x['br'] + (0.5 * x['to']) - x['fc'] - x['er'], axis=1)
df_ouro['as_rat'] = df_ouro.apply(lambda x: 100 * x['as'] / x['fga'] + (0.475 * x['llt']) + x['as'] + x['er'], axis=1)
df_ouro['er_rat'] = df_ouro.apply(lambda x: 100 * x['er'] / x['fga'] + (0.475 * x['llt'])  + x['as'] + x['er'], axis=1)
df_ouro['true_shot'] = df_ouro.apply(lambda x: 100 * x['pts'] / (2* x['fga'] + (0.475 * x['llt'])), axis=1)
df_ouro['eff_fg'] = df_ouro.apply(lambda x: (x['fgm'] + 0.5 * x['3pc']) / x['fga'], axis=1)
df_ouro['ll_rat'] = df_ouro.apply(lambda x: x['llt'] / x['fga'], axis=1)

# criando um data frame de Estatísticas Avançadas
adv_stats = df_ouro.filter(['Id','equipe', 'ef', 'vit', 'der', 'fga', 'fgm', 'poss', 'off_rtg', 'pie', 'as_rat', 'er_rat', 'true_shot', 'eff_fg', 'll_rat'], axis=1)
print('ESTATÍSTICAS AVANÇADAS - LIGA OURO')
adv_stats.head(8)
```

```
# criando Estatísticas Avançadas - NBB
df_nbb['fga'] = df_nbb.apply(lambda x: x['2pt'] + x['3pt'], axis=1)
df_nbb['fgm'] = df_nbb.apply(lambda x: x['2pc'] + x['3pc'], axis=1)
df_nbb['poss'] = df_nbb.apply(lambda x: 0.96 * (x['2pt'] + x['er'] + (0.45 * x['llt'] - x['ro'])), axis=1)
df_nbb['off_rtg'] = df_nbb.apply(lambda x: 100 * (x['pts'] / x['poss']), axis=1)
df_nbb['pie'] = df_nbb.apply(lambda x: x['pts'] + x['fgm'] + x['llc'] - x['fga'] - x['llt'] + x['rt'] + (0.5 * x['ro']) + x['as'] + x['br'] + (0.5 * x['to']) - x['fc'] - x['er'], axis=1)
df_nbb['as_rat'] = df_nbb.apply(lambda x: 100 * x['as'] / x['fga'] + (0.475 * x['llt']) + x['as'] + x['er'], axis=1)
df_nbb['er_rat'] = df_nbb.apply(lambda x: 100 * x['er'] / x['fga'] + (0.475 * x['llt'])  + x['as'] + x['er'], axis=1)
df_nbb['true_shot'] = df_nbb.apply(lambda x: 100 * x['pts'] / (2* x['fga'] + (0.475 * x['llt'])), axis=1)
df_nbb['eff_fg'] = df_nbb.apply(lambda x: (x['fgm'] + 0.5 * x['3pc']) / x['fga'], axis=1)
df_nbb['ll_rat'] = df_nbb.apply(lambda x: x['llt'] / x['fga'], axis=1)

# criando um data frame de Estatísticas Avançadas
adv_nbb = df_nbb.filter(['Id','equipe', 'ef', 'vit', 'der', 'fga', 'fgm', 'poss', 'off_rtg', 'pie', 'as_rat', 'er_rat', 'true_shot', 'eff_fg', 'll_rat'], axis=1)
print('ESTATÍSTICAS AVANÇADAS - NBB')
adv_nbb.head(16)
```

## Modelos preditivos de Machine Learning

*Machine Learning* ou aprendizado de máquina, é uma técnica que abrange a ideia de máquinas com a capacidade de aprenderem sozinhas a partir de grandes volumes de dados. Isso é feito por meio de algoritmos, identificando padrões de dados e criando conexões entre eles para aprender a executar uma tarefa sem a ajuda humana e de forma inteligente.

Esses algoritmos usam análises estatísticas para prever respostas mais precisamente e entregam o melhor resultado preditivo com menos chance de erro.

Essa tecnologia pode ser separada em duas categorias principais: supervisionada ou não supervisionada.

Os algoritmos de nosso interesse são os supervisionados, que são os que o ser humano precisa interagir controlando a saída e entrada de dados e interferindo no treinamento da máquina fazendo comentários sobre a precisão das previsões. Por fim, a máquina aplica o que foi aprendido no seu algoritmo para a próxima análise.

As vantagens de fazer isso, além de prover mais dados de maior qualidade para aumentar a acurácia do nosso modelo, é, no nosso caso, poder analisar essas variáveis que seriam "únicas e exclusivas", o que gera vantagem sobre times que não fazem esse tipo de trabalho.
